# Feriados no Brasil

```python
import requests
import json
import pandas as pd

# Para o ano 2025
url = "https://brasilapi.com.br/api/feriados/v1/2025"

payload = {}
headers = {
  'Accept': 'application/json'
}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```

```python
response_json = json.loads(response.text)
response_json
```

```python
df_holidays = pd.DataFrame(response_json)
df_holidays
```
